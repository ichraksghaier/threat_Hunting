```python
import re

def clean_and_correct_kql(kql_query):

    # Replace specific terms
    kql_query = kql_query.replace("Timestamp", "TimeGenerated")
    kql_query = kql_query.replace("DeviceDomain", "AccountDomain")

    # Remove redundant spaces and line breaks
    kql_query = re.sub(r'\s+', ' ', kql_query).strip()

    # Correct common KQL syntax issues
    kql_query = re.sub(r'\bwhere\b\s*\bwhere\b', 'where', kql_query, flags=re.IGNORECASE)
    kql_query = re.sub(r'\bsummarize\b\s*\bsummarize\b', 'summarize', kql_query, flags=re.IGNORECASE)
    kql_query = re.sub(r'\bproject\b\s*\bproject\b', 'project', kql_query, flags=re.IGNORECASE)

    # Ensure consistent casing for KQL keywords (optional)
    keywords = ["where", "project", "summarize", "join", "inner", "outer", "left", "right", "distinct", "order", "by"]
    for keyword in keywords:
        kql_query = re.sub(fr'\b{keyword}\b', keyword.upper(), kql_query, flags=re.IGNORECASE)

    # Additional corrections for known issues
    # Ensure logical conditions are correctly grouped
    kql_query = re.sub(r'\bor\b\s*\bor\b', 'or', kql_query, flags=re.IGNORECASE)
    kql_query = re.sub(r'\band\b\s*\band\b', 'and', kql_query, flags=re.IGNORECASE)

    # Correct the common time window expressions
    kql_query = re.sub(r'ago\(\d+d\)', lambda x: x.group().replace("d", " days"), kql_query)


    # Add color coding for KQL keywords (optional)
    colors = {
        'where': '\033[94m',    # Blue
        'project': '\033[92m',  # Green
        'summarize': '\033[93m',# Yellow
        'join': '\033[91m',     # Red
        'inner': '\033[95m',    # Magenta
        'outer': '\033[95m',    # Magenta
        'left': '\033[95m',     # Magenta
        'right': '\033[95m',    # Magenta
        'distinct': '\033[96m', # Cyan
        'order': '\033[97m',    # White
        'by': '\033[97m',       # White
        'reset': '\033[0m'      # Reset color
    }

    for keyword, color in colors.items():
        kql_query = re.sub(fr'\b{keyword.upper()}\b', f'{color}{keyword.upper()}{colors["reset"]}', kql_query, flags=re.IGNORECASE)

    # Format the query for readability
    formatted_query = re.sub(r'\s*\|\s*', r' |\n  ', kql_query)  # Add newlines before and after pipes and indent
    formatted_query = re.sub(r'(\bWHERE\b|\bPROJECT\b|\bSUMMARIZE\b|\bJOIN\b)', r'\n\1', formatted_query, flags=re.IGNORECASE)  # Add newlines before major keywords
    formatted_query = re.sub(r' {2,}', ' ', formatted_query)  # Reduce multiple spaces to single space
    formatted_query = formatted_query.replace('\n ', '\n')  # Remove leading spaces after newlines

    return formatted_query
```

```python
from rouge_score import rouge_scorer

def calculate_rouge_scores(references, predictions):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

    for ref, pred in zip(references, predictions):
        score = scorer.score(ref, pred)
        scores['rouge1'].append(score['rouge1'].fmeasure)
        scores['rouge2'].append(score['rouge2'].fmeasure)
        scores['rougeL'].append(score['rougeL'].fmeasure)

    avg_scores = {metric: sum(values) / len(values) for metric, values in scores.items()}
    return avg_scores

def evaluate_model(references, predictions):
    scores = calculate_rouge_scores(references, predictions)
    print(f"ROUGE-1 F1 Score: {scores['rouge1']:.4f}")
    print(f"ROUGE-2 F1 Score: {scores['rouge2']:.4f}")
    print(f"ROUGE-L F1 Score: {scores['rougeL']:.4f}")

# Example usage:
references = [
    "DeviceNetworkEvents| where isnotempty(RemoteIP)| where RemoteIP !in ('::1','-', '0.0.0.0') and RemoteIP !startswith '127.'| where TimeGenerated > ago(30d)| where isnotempty(AccountDomain)| summarize arg_max(TimeGenerated, DeviceName) by RemoteIP| project DeviceId, DeviceName, TimeGenerated, AccountDomain, RemoteIP"]

predictions = [
    "DeviceNetworkEvents| where isnotempty(RemoteIP)| where RemoteIP !in ('::1','-', '0.0.0.0') and not(RemoteIP startswith '127.')| where Timestamp > ago(30d)| where isnotempty(DeviceDomain)| summarize arg_max(Timestamp, *) by RemoteIP| project DeviceId, DeviceName, Timestamp, DeviceDomain, RemoteIP"]

evaluate_model(references, predictions)
```

```python
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.translate.meteor_score import meteor_score
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import numpy as np

# Functions to calculate ROUGE scores
def calculate_rouge_scores(references, predictions):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

    for ref, pred in zip(references, predictions):
        score = scorer.score(ref, pred)
        scores['rouge1'].append(score['rouge1'].fmeasure)
        scores['rouge2'].append(score['rouge2'].fmeasure)
        scores['rougeL'].append(score['rougeL'].fmeasure)

    avg_scores = {metric: sum(values) / len(values) for metric, values in scores.items()}
    return avg_scores

def evaluate_model_rouge(references, predictions):
    scores = calculate_rouge_scores(references, predictions)
    print(f"ROUGE-1 F1 Score: {scores['rouge1']:.4f}")
    print(f"ROUGE-2 F1 Score: {scores['rouge2']:.4f}")
    print(f"ROUGE-L F1 Score: {scores['rougeL']:.4f}")
    return scores

# Functions to calculate BLEU scores
def calculate_bleu_scores(references, predictions):
    bleu_scores = []
    for ref, pred in zip(references, predictions):
        ref_tokens = [ref.split()]
        pred_tokens = pred.split()
        score = sentence_bleu(ref_tokens, pred_tokens, smoothing_function=SmoothingFunction().method1)
        bleu_scores.append(score)
    avg_bleu = sum(bleu_scores) / len(bleu_scores)
    return avg_bleu

def evaluate_model_bleu(references, predictions):
    bleu_score = calculate_bleu_scores(references, predictions)
    print(f"BLEU Score: {bleu_score:.4f}")
    return bleu_score

# Functions to calculate Cosine Similarity scores
def calculate_cosine_similarity(references, predictions):
    vectorizer = TfidfVectorizer().fit_transform(references + predictions)
    vectors = vectorizer.toarray()
    ref_vectors = vectors[:len(references)]
    pred_vectors = vectors[len(references):]

    cosine_similarities = []
    for ref_vec, pred_vec in zip(ref_vectors, pred_vectors):
        cosine_sim = cosine_similarity([ref_vec], [pred_vec])[0][0]
        cosine_similarities.append(cosine_sim)

    avg_cosine_similarity = sum(cosine_similarities) / len(cosine_similarities)
    return avg_cosine_similarity

def evaluate_model_cosine(references, predictions):
    cosine_score = calculate_cosine_similarity(references, predictions)
    print(f"Cosine Similarity: {cosine_score:.4f}")
    return cosine_score

# Functions to calculate METEOR scores
def calculate_meteor_scores(references, predictions):
    meteor_scores = [meteor_score([ref.split()], pred.split()) for ref, pred in zip(references, predictions)]
    avg_meteor = sum(meteor_scores) / len(meteor_scores)
    return avg_meteor

def evaluate_model_meteor(references, predictions):
    meteor_score = calculate_meteor_scores(references, predictions)
    print(f"METEOR Score: {meteor_score:.4f}")
    return meteor_score

# Functions to calculate F1 scores
def calculate_f1_scores(references, predictions):
    ref_tokens = [ref.split() for ref in references]
    pred_tokens = [pred.split() for pred in predictions]

    f1_scores = []
    for ref, pred in zip(ref_tokens, pred_tokens):
        common = set(ref) & set(pred)
        if len(common) == 0:
            f1_scores.append(0.0)
        else:
            precision = len(common) / len(pred)
            recall = len(common) / len(ref)
            if precision + recall == 0:
                f1_scores.append(0.0)
            else:
                f1 = 2 * (precision * recall) / (precision + recall)
                f1_scores.append(f1)

    avg_f1 = sum(f1_scores) / len(f1_scores)
    return avg_f1

def evaluate_model_f1(references, predictions):
    f1_score = calculate_f1_scores(references, predictions)
    print(f"F1 Score: {f1_score:.4f}")
    return f1_score

# Visualization functions
def visualize_scores(scores):
    metrics = list(scores.keys())
    values = list(scores.values())

    plt.figure(figsize=(10, 5))
    plt.bar(metrics, values, color=['blue', 'green', 'red', 'orange', 'purple', 'cyan'])
    plt.xlabel('Metrics')
    plt.ylabel('Scores')
    plt.title('Evaluation Metrics Scores')
    plt.ylim(0, 1)
    plt.show()

def plot_scores_over_time(scores_dict):
    plt.figure(figsize=(10, 5))
    for metric, scores in scores_dict.items():
        plt.plot(range(len(scores)), scores, label=metric)
    plt.xlabel('Steps')
    plt.ylabel('Scores')
    plt.title('Evaluation Metrics Over Time')
    plt.legend()
    plt.show()

def plot_histogram_of_scores(scores):
    metrics = list(scores.keys())
    values = list(scores.values())

    plt.figure(figsize=(10, 5))
    plt.hist(values, bins=10, color='purple', alpha=0.7)
    plt.xlabel('Scores')
    plt.ylabel('Frequency')
    plt.title('Distribution of Evaluation Metrics Scores')
    plt.show()

# Example usage
references = [
    "DeviceNetworkEvents| where isnotempty(RemoteIP)| where RemoteIP !in ('::1','-', '0.0.0.0') and RemoteIP !startswith '127.'| where TimeGenerated > ago(30d)| where isnotempty(AccountDomain)| summarize arg_max(TimeGenerated, DeviceName) by RemoteIP| project DeviceId, DeviceName, TimeGenerated, AccountDomain, RemoteIP"]

predictions = [
    "DeviceNetworkEvents| where isnotempty(RemoteIP)| where RemoteIP !in ('::1','-', '0.0.0.0') and not(RemoteIP startswith '127.')| where Timestamp > ago(30d)| where isnotempty(DeviceDomain)| summarize arg_max(Timestamp, *) by RemoteIP| project DeviceId, DeviceName, Timestamp, DeviceDomain, RemoteIP"]

# Evaluate and visualize
rouge_scores = evaluate_model_rouge(references, predictions)
bleu_score = evaluate_model_bleu(references, predictions)
cosine_score = evaluate_model_cosine(references, predictions)
meteor_score = evaluate_model_meteor(references, predictions)
f1_score = evaluate_model_f1(references, predictions)

# Combine scores for visualization
combined_scores = {**rouge_scores, 'bleu': bleu_score, 'cosine_similarity': cosine_score, 'meteor': meteor_score, 'f1': f1_score}

# Visualize scores
visualize_scores(combined_scores)
```
